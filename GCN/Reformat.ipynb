```python
import os
import cv2
import glob
import numpy as np
import pandas as pd
import tqdm.notebook as tqdm
import matplotlib.pyplot as plt 
from multiprocessing.pool import Pool
```

```python
def create_folder(folder_format): 
    try: 
        os.makedirs(folder_format, exist_ok=True)
        print("[SUCCESS] The new directory is created!")
    except:
        print('[WARNING] The directory has been existed - Change for directory')
    pass

root = '/content/drive/MyDrive/COURSES/CS331/FinalProject/'
des_format = os.path.join(root, 'dataset', 'GraphCXR', 'CXR_Reformat') 
src_org = os.path.join(root, 'dataset', 'GraphCXR', 'CXR') 

create_folder(des_format)
```

```python
def Reformat(src, des): 
    # print('===From:', src,' ==to==', des,'===') 
    image = cv2.imread(src, cv2.IMREAD_GRAYSCALE)  
    cv2.imwrite(des, image)      
```

```python
for category in tqdm.tqdm(os.listdir(os.path.join(src_org, 'test'))): 
    folder_category = os.path.join(src_org, 'test', category)

    path_to = os.path.join(des_format, 'test', category) 
    create_folder(path_to)

    for img in os.listdir(folder_category):   
        Reformat(
            src=os.path.join(folder_category, img), 
            des=os.path.join(os.path.join(path_to, img)))
```

```python
file_collection = []
```

```python
for category in os.listdir(os.path.join(src_org, 'train')): 
    folder_category = os.path.join(src_org, 'train', category) 
    for img in os.listdir(folder_category):  
        # print('Src:', os.path.join(folder_category, img))  
        file_collection.append(os.path.join(folder_category, img)) 

for category in os.listdir(os.path.join(src_org, 'val')): 
    folder_category = os.path.join(src_org, 'val', category) 
    for img in os.listdir(folder_category):  
        # print('Src:', os.path.join(folder_category, img))  
        file_collection.append(os.path.join(folder_category, img)) 
```

```python
from sklearn.model_selection import train_test_split
train, val = train_test_split(file_collection, test_size=0.2, shuffle=True, random_state=42)
```

```python
len(train), len(val)
```

```python
cnt = 0
create_folder(os.path.join(des_format, 'train', 'normal'))
create_folder(os.path.join(des_format, 'train', 'virus'))
create_folder(os.path.join(des_format, 'train', 'bacteria'))
for file in tqdm.tqdm(train): 
    if file.split('/')[-2] == 'normal':  
        Reformat(
            src=file, 
            des=os.path.join(os.path.join(des_format, 'train', 'normal', 'img-{}.jpg'.format(cnt))))
    
    elif file.split('/')[-2] == 'virus':  
        Reformat(
            src=file, 
            des=os.path.join(os.path.join(des_format, 'train', 'virus', 'img-{}.jpg'.format(cnt))))
        
    elif file.split('/')[-2] == 'bacteria':  
        Reformat(
            src=file, 
            des=os.path.join(os.path.join(des_format, 'train', 'bacteria', 'img-{}.jpg'.format(cnt))))

    cnt += 1
```

```python
cnt = 0
create_folder(os.path.join(des_format, 'val', 'normal'))
create_folder(os.path.join(des_format, 'val', 'virus'))
create_folder(os.path.join(des_format, 'val', 'bacteria'))
for file in tqdm.tqdm(val): 
    if file.split('/')[-2] == 'normal':  
        Reformat(
            src=file, 
            des=os.path.join(os.path.join(des_format, 'val', 'normal', 'img-{}.jpg'.format(cnt))))
    
    elif file.split('/')[-2] == 'virus':  
        Reformat(
            src=file, 
            des=os.path.join(os.path.join(des_format, 'val', 'virus', 'img-{}.jpg'.format(cnt))))
        
    elif file.split('/')[-2] == 'bacteria':  
        Reformat(
            src=file, 
            des=os.path.join(os.path.join(des_format, 'val', 'bacteria', 'img-{}.jpg'.format(cnt))))

    cnt += 1
```
